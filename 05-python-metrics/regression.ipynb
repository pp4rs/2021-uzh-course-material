```python
import pandas as pd
```

```python
df1 = pd.read_stata('data/maketable1.dta')
```

```python
df1.head()
```

```python
df1.describe()
```

```python
import statsmodels.formula.api as smf
```

```python
simple_reg = smf.ols('logpgp95 ~ avexpr', data = df1)
```

```python
res_simple_reg = simple_reg.fit()
```

```python
res_simple_reg.summary()
```

```python
df2 = pd.read_stata('data/maketable2.dta')
```

```python
df2.head()
```

```python
reg2 = smf.ols('logpgp95  ~ avexpr + lat_abst', data = df2)
```

```python
res_reg2 = reg2.fit()
```

```python
res_reg2.summary()
```

Can we put these results in a table??

```python
from stargazer.stargazer import Stargazer
```

```python
tab = Stargazer([res_simple_reg, res_reg2])
```

```python
Stargazer([res_simple_reg, res_reg2])
```

```python
reg3 = smf.ols('logpgp95 ~ avexpr + lat_abst + asia + africa + other', data = df2)
```

```python
res_reg3  = reg3.fit()
```

```python
res_reg3.summary()
```

```python
# robust standard errors

reg3_robust = reg3.fit(cov_type= 'HC3')
```

```python
reg3_robust.summary()
```

```python
Stargazer([res_reg3, reg3_robust])
```

# IV regression

```python
from linearmodels.iv import IV2SLS
```

```python
df4 = pd.read_stata('data/maketable4.dta')
```

```python
df4.head()
```

```python
df4 = df4.query('baseco == 1')
```

```python
iv_reg = IV2SLS.from_formula('logpgp95 ~ 1 + [avexpr ~ logem4] + lat_abst + asia + africa', data = df4)
```

```python
res_iv_reg = iv_reg.fit()
```

```python
res_iv_reg.summary
```

```python
tab = Stargazer([res_simple_reg, res_reg2, res_reg3])

with open('my_results.tex', 'w') as f:
    f.write(tab.render_latex())
```
