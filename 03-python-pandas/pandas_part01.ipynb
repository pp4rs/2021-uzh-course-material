# Importing many data sets

```python
import glob
```

```python
data_path = 'pandas-data/data/bls-employment-state/*'
bls_files = glob.glob(data_path)
```

```python
ue_files = glob.glob(data_path + '3.csv')
```

```python
ue_files
```

```python
cols = ['state', 
        'year', 
        'period', 
        'value']

# read in all the files in the list ue_list
unemp_data = (pd.read_csv(iFile, 
                          usecols= cols) 
                for iFile in ue_files)
```

```python
unemp_data
```

```python
unemployment = pd.concat(unemp_data)
```

```python
n_unemp_files = glob.glob(data_path + "4.csv")
employed_files = glob.glob(data_path + "5.csv")
labourforce_files = glob.glob(data_path + "6.csv")
```

```python
n_unemp_data = (pd.read_csv(iFile, 
                          usecols= cols) 
                for iFile in n_unemp_files)
employed_data = (pd.read_csv(iFile, 
                          usecols= cols) 
                for iFile in employed_files)
labourforce_data = (pd.read_csv(iFile, 
                          usecols= cols) 
                for iFile in labourforce_files)
```

```python
n_unemployed = pd.concat(n_unemp_data)
employed = pd.concat(employed_data)
labourforce = pd.concat(labourforce_data)
```

```python
# alternative way
def append_data(data_path, my_string, col_list):

    files = glob.glob(data_path + my_string)

    data_gen = (pd.read_csv(iFile, usecols = col_list)
                for iFile in files
                )
    return pd.concat(data_gen)
```

```python
n_unemp_files = glob.glob(data_path + "4.csv")
employed_files = glob.glob(data_path + "5.csv")
labourforce_files = glob.glob(data_path + "6.csv")
```

```python
n_unemployed = append_data(data_path = data_path, my_string = '4.csv', col_list = cols)
```

```python
unemployment.rename(columns = {'value': 'ue_rate'},
                    inplace = True)
n_unemployed.rename(columns = {'value': 'ue_qty'},
                    inplace = True)
employed.rename(columns = {'value': 'empl_qty'},
                    inplace = True)
labourforce.rename(columns = {'value': 'lf_qty'},
                    inplace = True)
```

```python
unemployment.set_index(['state', 'year', 'period'], inplace= True)
```

```python
n_unemployed.set_index(['state', 'year', 'period'], inplace = True)
```

```python
n_unemployed.head()
```

```python
merged_data = unemployment.merge(n_unemployed,
                                    left_index= True, 
                                    right_index= True,
                                    how = 'inner'
                                )
```

```python
employed.set_index(['state', 'year', 'period'], inplace= True)
labourforce.set_index(['state', 'year', 'period'], inplace= True)
```

```python
merged_data = merged_data.merge(employed, 
                                left_index = True, 
                                right_index = True,
                                how =  'inner'
)
```

```python
merged_data = merged_data.merge(labourforce, 
                                left_index = True, 
                                right_index = True,
                                how =  'inner'
)
```

```python
merged_data
```

```python
# hmmm can we merge all that data better (less clunkily)

dfs = [unemployment, n_unemployed, employed, labourforce]
```

```python
from functools import reduce
```

```python
merged_data_2 = reduce(lambda df1, df2: pd.merge(df1, df2, 
                        left_index= True, 
                        right_index= True, 
                        how = 'inner'
                        ), 
        dfs 
)
```
