Create a function "my_query(dataset, year_start, year_end, state)" that returns the correct subset of data as a non copy

if you are feeling lucky, try and return the output sorted by year, month

```python
def my_query(dataset, year_start, year_end, state, sort_by):

    out = dataset.query('@year_end >= year >= @year_start & state ==@state')\
          .copy()
    # heres a comment
    out = out\
          .sort_values(by = sort_by)\
          .reset_index(drop = True)  
    
    return out
```
